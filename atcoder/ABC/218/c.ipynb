```python
n = int(input())

s = []

for _ in range(n):
    tmp = list(input())
    s.append(tmp)
    
t = []

for _ in range(n):
    tmp = list(input())
    t.append(tmp)
    


def cut_grid(grid):
    n = len(grid)
    max_x = -1
    max_y = -1
    min_x = n + 10
    min_y = n + 10
    # 図形を切り出す
    for i in range(n):
        for j in range(n):
            if grid[i][j] == "#":
                max_x = max(max_x, j)
                max_y = max(max_y, i)
                min_x = min(min_x, j)
                min_y = min(min_y, i)

    x = [min_x, max_x]
    y = [min_y, max_y]
    
    new_grid = []
    for i in range(y[0], y[1] + 1):
        tmp = []
        for j in range(x[0], x[1] + 1):
            tmp.append(grid[i][j])
        new_grid.append(tmp)

    return new_grid

def rotate_grid(grid):
    n = len(grid)
    m = len(grid[0])
    rotated_grid = [[0] * n for _ in range(m)]
    for i in range(n):
        for j in range(m):
            rotated_grid[j][n - 1 - i] = grid[i][j]
    return rotated_grid

def is_equal(grid1, grid2):
    if len(grid1) != len(grid2) or len(grid1[0]) != len(grid2[0]):
        return False
    for i in range(len(grid1)):
        for j in range(len(grid1[0])):
            if grid1[i][j] != grid2[i][j]:
                return False
    return True

def check_rotation(grid1, grid2):
    for _ in range(4):
        if is_equal(grid1, grid2):
            return True
        grid1 = rotate_grid(grid1)
    return False


new_s = cut_grid(s)
new_t = cut_grid(t)



result = check_rotation(new_s, new_t)
if result:
    print("Yes")
else:
    print("No")
```
