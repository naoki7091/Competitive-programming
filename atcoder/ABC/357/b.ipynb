```python
s = list(input())
cnt_C = 0
cont_c = 0
for ch in s:
    if 65 <= ord(ch) <= 90:
        cnt_C += 1
    else:
        cont_c += 1
ans = []
# 大文字にする
if cont_c < cnt_C:
    for ch in s:
        if 97 <= ord(ch) <= 122:
            ans.append(chr(ord(ch) - 32))
        else:
            ans.append(ch)
else:
    for ch in s:
        if 65 <= ord(ch) <= 90:
            ans.append(chr(ord(ch) + 32))
        else:
            ans.append(ch)

print(*ans,sep="")
```
