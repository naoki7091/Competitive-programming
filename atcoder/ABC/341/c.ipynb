```python
y ,x , n = map(int,input().split())

t = list(input())

grid = []

for _ in range(y):
    tmp = list(input())
    grid.append(tmp)





def dfs(i, j, t, index):
    if index == len(t):
        return True
    
    move_ch = t[index]
    if move_ch == "L":
        next_i, next_j = i, j - 1
    elif move_ch == "R":
        next_i, next_j = i, j + 1
    elif move_ch == "U":
        next_i, next_j = i - 1, j
    elif move_ch == "D":
        next_i, next_j = i + 1, j
    
    if not (0 <= next_i < y and 0 <= next_j < x):
        return False
    if grid[next_i][next_j] == "#":
        return False

    return dfs(next_i, next_j, t, index + 1)

ans = 0
for i in range(y):
    for j in range(x):
        if grid[i][j] == "#":
            continue
        if dfs(i, j, t, 0):
            ans += 1

print(ans)
```
