```python
n = int(input())


def s(n : int) -> str:
    if n == 1:
        return [1]
    else:
        return s(n - 1) + [n] + s(n - 1)


print(*s(n))
```

```python
n = int(input())

s = []

def f(s_minus1 : str, tmp_n : str) -> str:
    return s_minus1 + tmp_n + s_minus1

for i in range(1, n + 1):
    s = f(s, [i])



print(*s)
```
