```python
n, m = map(int, input().split())
S = []

for _ in range(n):
    s = input()
    S.append(s)
    
min_count = n 
for bit in range(1, 1 << n):
    can_convert = True
    for j in range(m):
        if all(S[i][j] == 'x' for i in range(n) if bit & (1 << i)):
            can_convert = False
            break
    if can_convert:
        min_count = min(min_count, bin(bit).count('1'))


print(min_count)
```

```python
n, m = map(int, input().split())
S = []

for _ in range(n):
    s = input()
    S.append(s)


ans = 10**10
for bit in range(1<<n):
    tmp_s = []
    for i in range(n):
        if (1<<i) & bit:
            tmp_s.append(S[i])
    count = [0]*m
    for st in tmp_s:
        for i in range(m):
            if st[i] == "o":
                count[i] = "o"
    count_o = 0
    for ch in count:
        if ch == "o":
            count_o += 1
    if count_o == m:
        ans = min(ans,len(tmp_s))   

print(ans)
```
