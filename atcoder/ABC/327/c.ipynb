```python
a = []

for i in range(9):
    tmp = list(map(int,input().split()))
    a.append(tmp)


# query1:各行について、調べる
def colm(a):
    flag = True
    for i in range(len(a)):
        judge_set = set()
        for j in range(len(a)):
            judge_set.add(a[i][j])
        # print(judge_set)
        if len(judge_set) != len(a):
            flag = False
            return flag
    
    return flag

def row(a):
    flag = True
    for i in range(len(a)):
        judge_set = set()
        for j in range(len(a)):
            judge_set.add(a[j][i])
        if len(judge_set) != len(a):
            flag = False
            return flag
    return flag


def split_a(a):
    start_i = 0
    end_i = 3
    start_j = 0
    end_j = 3
    count = 0

    while count < 9:
        judge_set = set()
        for i in range(start_i, end_i):
            for j in range(start_j, end_j):
                judge_set.add(a[i][j])

        if len(judge_set) != 9:
            return False
        count += 1
        start_j += 3
        end_j += 3

        if count % 3 == 0:  # 各行の最後のブロック後にインデックスをリセット
            start_i += 3
            end_i += 3
            start_j = 0
            end_j = 3

        if count == 9:  # 全てのブロックを処理したらループを終了
            break
    
    return True

    

judge_1 = colm(a)
judge_2 = row(a)
judge_3 = split_a(a)

if judge_1 and judge_2 and judge_3:
    print("Yes")
else:
    print("No")
```
